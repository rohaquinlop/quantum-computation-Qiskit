## The Hadamard Gate

The Hadamard gate (H-gate) is a fundamental quantum gate. It allows us to move away from the poles of the Bloch sphere and create a superposition of |0> and |1>. It has the matrix:
```Python
H = 1/sqrt(2) [ [1, 1], [1, -1] ]
```

H|0> = |+> , H|1> = |->
